Seems to be a relatively normal distribution, with very few old people, and a low-to-moderate amount of children. Was there an increase in liklihood of survival for children, however?

##### Effect of Ticket Price on Survival

```python
training_df["Fare"].describe()
```

```python
plt.figure(figsize=(20,6))
sns.histplot(training_df["Fare"], kde=True)
plt.title("Fare Distribution")
plt.show()
```

This is not a normal distribution, with fares heavily skewed to the lower end. This means this is a good candidate for binning these, and then comparing them to the survival rate. Lets do that and see what happens

```python
fare_bucket_names = ["Cheap", "Normal", "Pricey", "Expensive"]

fare_ranges = pd.qcut(training_df["Fare"], len(fare_bucket_names), labels=fare_bucket_names)

plt.figure(figsize=(20,6))
sns.barplot(x = fare_ranges, y = training_df["Survived"])
plt.title("Survival Rate by Fare bucket")
plt.show()
```

As expected, how much you paid for your ticket is directly proportional to how likely you were to survive. What if we binned Age as well?

```python
age_bucket_names = ["Baby", "Kid", "Young Adult", "Adult", "Middle Aged", "Elderly"]

age_ranges = pd.qcut(training_df["Age"], len(age_bucket_names), labels=age_bucket_names)

plt.figure(figsize=(20,6))
sns.barplot(x = age_ranges, y = training_df["Survived"])
plt.title("Survival Rate by Age bucket")
plt.show()
```

Binning the age groups does not seem to make a huge difference as it pertains to the survival rate. However, it is clear that the youngest of children had the highest survival rate.

### Feature Engineering

```python
features_to_use = ["Pclass", "Sex", "Age", "Fare"]

training_df.head()
```

```python
def create_submission(classifier, X, y, X_test):
    # Fit the passed-in classifier
    classifier.fit(X, y)
    # Run predictions
    y_pred = classifier.predict(X_test)

    # Print the dataframe to a csv in the same location as this notebook
    output = pd.DataFrame({'PassengerId': testing_df.PassengerId, 'Survived': y_pred})
    output.to_csv('submission.csv', index=False)
    print("Submission saved.")
```

```python
# Function from your lecture which runs a classifier and reports the results
def evaluate_classifier(classifier, X, y, iterations, text=''):
    accs = []
    for i in range(iterations):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i, stratify=y)
        
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict(X_test)
        accs += [accuracy_score(y_test, y_pred)]

    print(f'{text:<20s} acc={np.mean(accs):.3f} with {iterations} iterations')
```

```python
numerical_cols = ["Fare", "Age"]
categorical_cols = ["Pclass"]

numerical_transformer = SimpleImputer(strategy="median")

# Trying out the sklearn "Pipeline" from the book here:
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('onehot', OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical_cols),
    ('cat', categorical_transformer, categorical_cols)
])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', GradientBoostingClassifier(max_depth=5, random_state=1))
])

pipeline2 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1))
])

pipeline3 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic', random_state=1))
])

X = training_df[features_to_use]
y = training_df["Survived"]

pipeline.fit(X, y)

print(f'Cross val score: {cross_val_score(pipeline, X, y, cv=10).mean()}')

print(f'Cross val score: {cross_val_score(pipeline2, X, y, cv=10).mean()}')

print(f'Cross val score: {cross_val_score(pipeline3, X, y, cv=10).mean()}')

create_submission(pipeline, X, y, testing_df[features_to_use])
```

After imputing and encoding values, for some reason my pipeline performs worse than just one-hot encoding values and otherwise using the dataset as-is, which you can find in the section below.

```python
features_to_use = ["Pclass", "Sex", "SibSp", "Parch"]

X = pd.get_dummies(training_df[features_to_use])
y = training_df["Survived"]

X_test = pd.get_dummies(testing_df[features_to_use])

# Testing the performance of a bunch of classifiers locally using stratified 80/20 splits
evaluate_classifier(svm.SVC(class_weight='balanced', kernel='linear', C=2), X, y, 10, 'SVM (linear)')
evaluate_classifier(svm.SVC(class_weight='balanced', kernel='rbf', gamma=2, C=2), X, y, 10, 'SVM (rbf)')

print('--------------')

evaluate_classifier(XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic', random_state=1), X, y, 10, 'XGBoost')
evaluate_classifier(GradientBoostingClassifier(n_estimators=100, max_depth=5, random_state=1), X, y, 10, 'GradientBoosting')
evaluate_classifier(HistGradientBoostingClassifier(random_state=1), X, y, 10, 'HistGradientBoosting')

print('--------------')

evaluate_classifier(GaussianNB(), X, y, 10, 'GaussianNB')

print('--------------')

evaluate_classifier(KNeighborsClassifier(n_neighbors=3), X, y, 10, 'KNeighbors')

print('--------------')

evaluate_classifier(AdaBoostClassifier(n_estimators=100, random_state=1), X, y, 10, 'AdaBoost')

print('--------------')

evaluate_classifier(ExtraTreesClassifier(n_estimators=100, random_state=1), X, y, 10, 'ExtraTrees')
evaluate_classifier(DecisionTreeClassifier(random_state=1), X, y, 10, 'DecisionTree')
evaluate_classifier(RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1), X, y, 10, 'RandomForest')
```
